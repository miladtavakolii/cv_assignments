```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Load image and convert to grayscale

```python
image = cv2.imread("images/Lenna.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

# 2. Thresholding to create a noisy binary image

```python
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

# 3. Define structuring element (kernel)

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))  # You can use (5,5)
```

# 4. Apply morphological cleaning

```python
# Opening: Remove small white noise (salt noise)
opened = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

# Closing: Fill small black holes (pepper noise)
cleaned = cv2.morphologyEx(opened, cv2.MORPH_CLOSE, kernel)
```

# 5. Display results

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Noisy Binary Image")
plt.imshow(binary, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Cleaned Image (Opening + Closing)")
plt.imshow(cleaned, cmap="gray")
plt.axis("off")

plt.show()
```
